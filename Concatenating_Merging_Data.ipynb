## Merge in Pandas

Let's have two Scenario of "Customer sales" and "Customer Info" and let's merge both the tables using Merge

```python
# Customer Sales Month wise given below - 
jan_purchases = pd.DataFrame({
    'Customer_ID': [1, 2, 3],
    'Product': ['Shoes', 'Bag', 'Hat'],
    'Amount': [1000, 500, 105]
})

feb_purchases = pd.DataFrame({
    'Customer_ID': [1, 4, 5],
    'Product': ['Socks', 'Shirt', 'Shoes'],
    'Amount': [75, 700, 1200]
})


all_purchases = pd.concat([jan_purchases, feb_purchases], axis=0, ignore_index=True)
print("All Purchases (Vertical):\n", all_purchases)
```

Now we have another table - Customer Details we need to merge them.


```python
customer_details = pd.DataFrame({
    'Customer_ID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'],
    'Age': [25, 32, 40, 29, 22],
    'Location': ['NY', 'CA', 'TX', 'NV', 'FL']
})
```

```python
customer_details
```

```python
customer_merged_data = pd.merge(all_purchases, customer_details, on ='Customer_ID', how = 'inner')
```

```python
customer_merged_data
```

Scenario's for Merge Data - 
Inner Join (Only Matching Records)
Left Join (All from Left, Matches from Right)
Right Join (All from Right, Matches from Left)
Outer Join (All Records from Both Sides)

```python
sales_record = DataFrame({"customer_id":[1,2,3,4,5,6,7,8,9],
                          "product_name":['Shoes','socks','Shoes','socks','Shoes',
                                'socks','Shoes','socks','Shoes'],
                           "product_price":[1200,120,1200,120,1200,120,1200,120,1200]     
                                })
```

```python
print(sales_record)
```

```python
cust_detail = DataFrame({
                "customer_unique_id":[1,3,5,7,9,11,13,15,17],
                "customer_name":["Ajay","Bjay","Cjay","Djay","Ejay","Fjay","Jjay","Hjay","Ijay"],
                "cust_address":["AZ","BZ","CZ","DZ","EZ","FZ","JZ","HZ","IZ"]})
```

```python
cust_detail
```

```python
inner_merge = pd.merge(sales_record,cust_detail,left_on='customer_id', right_on="customer_unique_id", how="inner")
```

```python
print(inner_merge)
```

```python
left_merge = pd.merge(sales_record,cust_detail,left_on='customer_id', right_on="customer_unique_id", how="left")
```

```python
print(left_merge)
```

```python
right_merge = pd.merge(sales_record,cust_detail,left_on='customer_id', right_on="customer_unique_id", how="right")
```

```python
print(right_merge)
```

```python
outer_merge = pd.merge(sales_record,cust_detail,left_on='customer_id', right_on="customer_unique_id", how="outer")
```

```python
print(outer_merge)
```
