```python
import pandas as pd
import numpy as np

from pandas import DataFrame
```

Filling missing values using fillna(), replace(), interpolate()

```python
data = {"names":['Ajay', 'Bijay', 'Cijay', 'Djay', 'Ejay', 'Fjay', 'Gjay', 'Hjay'],
        "age":[30,29,28,27,26,25,24,23],
        "gender":["MALE","MALE","MALE","MALE","MALE","MALE","MALE","MALE"],
        "rank":[2,1,4,6,7,8,3,5]
        }
```

```python
ranking_df = DataFrame(data)
ranking_df.iloc[2:5,1] = np.nan
ranking_df.iloc[3:6,3] = np.nan
ranking_df.iloc[3,:] = np.nan
print(ranking_df)
```

```python
ranking_df.isnull()
#False represents not missing values.
#True represents missing values.
```

```python
ranking_df.notnull()
#opposite of isull()
```

```python
ranking_df.isna()
# same as isnull()
```

```python
#isnull()
#notnull()
```

```python
bool_series = pd.isnull(ranking_df['age'])
```

```python
print(bool_series)
```

```python
print(ranking_df[bool_series])
```

```python
ranking_df.fillna(0)
```

```python
ranking_df.fillna(method='pad')
```

```python
ranking_df.fillna(method='bfill')
```

```python
# Method pad fill the Null value with the prior one.
# Method bfill fill the Null Value with the before ones value.
```

## Interpolate()
Fill it with the linear interpolation.

```python
ranking_df.interpolate(method='linear')
```

```python
ranking_df.dropna()
```
