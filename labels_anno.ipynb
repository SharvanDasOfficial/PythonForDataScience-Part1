```python
# The Object Oriented Method
```

```python
rcParams['figure.figsize'] = 8,4
mpg = cars_df.mpg
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])
mpg.plot()
ax.set_xticks(range(32))
ax.set_xticklabels(cars_df['types'], rotation = 60, fontsize = 'medium')
ax.legend(loc = 'best')
```

```python
rcParams['figure.figsize'] = 12,4
mpg = cars_df.mpg
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])
mpg.plot()
ax.set_xticks(range(32))
ax.set_xticklabels(cars_df['types'], rotation = 60, fontsize = 'medium')
ax.set_xlabel('car names')
ax.set_ylabel('mil / gal')
ax.set_title("Miles per Galoon")
ax.set_ylim([0,45])
ax.annotate('Totota Corola', xy = (19,33.9), xytext=(21,35), arrowprops=dict(facecolor = 'black', shrink = 0.01))
ax.annotate('Cadillac Fleetwood', xy = (15,10.4), xytext=(17,12), arrowprops=dict(facecolor = 'red', shrink = 0.05))

ax.legend(loc = 'best')
```
