```python
# Scraping book titles and prices from an example website.
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'http://books.toscrape.com/'
```

```python
print(url)
```

```python
requests.get(url)
```

```python
response = requests.get(url)
```

```python
print(response)
```

```python
print(response.content)
```

```python
# Parse the HTML
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
print(soup)
```

```python
books = soup.find_all('h3')
```

```python
for book in books:
    print(book.text)
```

```python
prices = soup.find_all('p',class_='price_color')
```

```python
for price in prices:
    print(price)
```

```python
for book, price in zip(books, prices):
    print(book.text, price.text)
```
